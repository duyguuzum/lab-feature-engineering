```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('learningSet.csv')
```

```python
print(data.shape)
data.head()
```

```python
numerical_columns = ['WEALTH1', 'ADI', 'DMA', 'MSA']
null_counts = data[numerical_columns].isnull().sum()
print(null_counts)
```

```python
data['GEOCODE2'].fillna('Unknown', inplace=True)
```

```python
data['WEALTH1'].fillna(data['WEALTH1'].median(), inplace=True)
data['ADI'].fillna(data['ADI'].mean(), inplace=True)
data['DMA'].fillna(data['DMA'].mean(), inplace=True)
data['MSA'].fillna(data['MSA'].mean(), inplace=True)
```

```python
import matplotlib.pyplot as plt

data[['WEALTH1', 'ADI', 'DMA', 'MSA']].hist(bins=20, figsize=(12, 6))
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = data[['WEALTH1', 'ADI', 'DMA', 'MSA']].corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 6))
data[['WEALTH1', 'ADI', 'DMA', 'MSA']].boxplot()
plt.show()
```
